### Динамика длины при низких температурах (0.05, 0.15,0.25)

```python
lines = open('leng_over_t_0.050000.txt').readlines()[0]
lines = lines.split()
length1 = []
for l in lines:
    length1.append(float(l))
    
lines = open('leng_over_t_0.150000.txt').readlines()[0]
lines = lines.split()
length2 = []
for l in lines:
    length2.append(float(l))
    

lines = open('leng_over_t_0.250000.txt').readlines()[0]
lines = lines.split()
length3 = []
for l in lines:
    length3.append(float(l))
```

```python
plt.figure(figsize=(10, 6))

plt.title("Длина цепочки с течением времени")
plt.plot([i for i in range(len(length1 ))], length1, ":", label = "0.05" )

plt.plot([i for i in range(len(length2 ))], length2 ,label = "0.15")

plt.plot([i for i in range(len(length3 ))], length3 ,label = "0.25" )
plt.xlabel("Номер итерации", fontsize = 14)
plt.ylabel("Длина", fontsize=14)
plt.legend(loc = "best", title = "Температура")
```
